```python
import pandas as pd

codes = pd.read_csv('codes.csv')
codes.head()
```

```python
contest513 = dict()

for i in range(len(codes)):
    id = codes.loc[i,'Unnamed: 0']
    student = codes.loc[i,'author_id']
    problem = codes.loc[i,'problem_id']
    status = codes.loc[i,'status_percent']
    path = codes.loc[i,'id']
    # status = (0 if codes.loc[i,'status_percent'] < 100 else 1)
    
    if contest513.get(student, None) == None:
        contest513[student] = list()
    contest513[student].append([int(id),int(problem),int(status),int(path)])

for key,value in contest513.items():
    value.sort()
print(contest513[1])
```

```python
import random
import csv
import numpy as np

train_file = open('contest513_train.csv', 'a+', encoding='utf-8', newline='')
test_file = open('contest513_test.csv', 'a+', encoding='utf-8', newline='')
train_writer = csv.writer(train_file)
test_writer = csv.writer(test_file)

for key,value in contest513.items():
    data = np.array(value)
    if random.random() < 0.8:
        train_writer.writerow([len(value)])
        train_writer.writerow(data[:,1])
        train_writer.writerow(data[:,2])
        train_writer.writerow(data[:,3])
    else:
        test_writer.writerow([len(value)])
        test_writer.writerow(data[:,1])
        test_writer.writerow(data[:,2])
        test_writer.writerow(data[:,3])
train_file.close()
test_file.close()
```
